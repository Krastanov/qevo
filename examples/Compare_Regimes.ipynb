```python
conf['P2'] = 0.99
conf['MÎ·'] = 0.99
p = Population(**conf)
p.run()
```

```python
c1 = p.l[0]
c1
```

```python
conf['P2'] = 0.95
conf['MÎ·'] = 0.95
p = Population(**conf)
p.run()
```

```python
c2 = p.l[0]
c2
```

```python
conf['P2'] = 0.98
conf['MÎ·'] = 0.98
p = Population(**conf)
p.run()
```

```python
c3 = p.l[0]
c3
```

```python
conf['P2'] = 0.999
conf['MÎ·'] = 0.999
p = Population(**conf)
p.run()
```

```python
c4 = p.l[0]
c4
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = np.logspace(-4,-1)
```

```python
calc = lambda c, e: c.copy(P2=1-e,MÎ·=1-e).fidelity()
```

```python
Fs = np.array([np.array([calc(c, _) for _ in eps]) for c in [c2,c3,c1,c4]])
```

```python
ps = [0.95, 0.98, 0.99, 0.999]
plt.close('all')
plt.figure()
plt.loglog(eps, 1-Fs.T)
plt.vlines(1-np.array(ps), np.min(1-Fs), np.max(1-Fs), alpha=0.3, linestyles='--')
plt.legend(ps)
plt.gcf()
```

```python
plt.close('all')
plt.figure()
plt.loglog(eps, 1-Fs.T[:,2:])
plt.gcf()
```
